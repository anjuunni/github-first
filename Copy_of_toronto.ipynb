```
toronto_neigh_df= pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
toronto_neigh_df.head()
#scraping data from the page
```

```
toronto_neigh_df.drop(toronto_neigh_df.loc[toronto_neigh_df['Borough']=="Not assigned"].index, inplace=True)#Removing raws having value not assigned.
```

```
toronto_neigh_df.head()
```

```
toronto_neigh_df.groupby(['Postcode','Borough'], sort = False).agg(lambda x: ','.join(x))#joining raws that have same postal code 
```

```
toronto_neigh_df.loc[toronto_neigh_df['Neighbourhood'] == "Not assigned", 'Neighbourhood'] = toronto_neigh_df["Borough"]#changing neighbourhood column value that have not assigned to borough value
```
